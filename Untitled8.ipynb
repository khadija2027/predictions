```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import root_mean_squared_error, mean_squared_error
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
import warnings
warnings.filterwarnings('ignore')

# -----------------------------
# 1️⃣ Charger les données
# -----------------------------
train = pd.read_csv("train_final_ml.csv")
test = pd.read_csv("test_final_ml.csv")

target = "Rendement_T_ha"
y = train[target]

# -----------------------------
# 2️⃣ Feature engineering AMÉLIORÉ
# -----------------------------
for df in [train, test]:
    df["Date Plantation"] = pd.to_datetime(df["Date Plantation"])
    df["Date Arrachage"] = pd.to_datetime(df["Date Arrachage"])

    df["Age_plant"] = ((df["Date Arrachage"] - df["Date Plantation"]).dt.days / 30.44).clip(0,240)
    df["Duree_Culture_Jours"] = (df["Date Arrachage"] - df["Date Plantation"]).dt.days
    df["Age_plant_sq"] = df["Age_plant"] ** 2
    df["Age_plant_cube"] = df["Age_plant"] ** 3  # Nouveau

    df["Total_NPK"] = (
        df["Apport_N_(kg/ha)"] +
        df["Apport_P_(kg/ha)"] +
        df["Apport_K_(kg/ha)"]
    )

    # Nouvelles features climat
    df["Temp_Amplitude"] = df["TemperatureMax C°"] - df["TemperatureMin C°"]
    df["Temp_Mean"] = (df["TemperatureMax C°"] + df["TemperatureMin C°"]) / 2
    df["Temp_NPK"] = df["Temp_Mean"] * df["Total_NPK"]

    # Ratios NPK
    df["N_P_ratio"] = df["Apport_N_(kg/ha)"] / (df["Apport_P_(kg/ha)"] + 1)
    df["N_K_ratio"] = df["Apport_N_(kg/ha)"] / (df["Apport_K_(kg/ha)"] + 1)
    df["P_K_ratio"] = df["Apport_P_(kg/ha)"] / (df["Apport_K_(kg/ha)"] + 1)

    # Densité nutritionnelle
    df["NPK_per_day"] = df["Total_NPK"] / (df["Duree_Culture_Jours"] + 1)

    # Age optimal
    df["Age_deviation"] = np.abs(df["Age_plant"] - 180)

    if "Irrigation" in df.columns:
        df["Irrigation_times_Age"] = df["Irrigation"] * df["Age_plant"]
        df["Irrigation_NPK"] = df["Irrigation"] * df["Total_NPK"]

# -----------------------------
# 3️⃣ Suppression des colonnes inutiles
# -----------------------------
drop_cols = [
    "Date Plantation", "Date Arrachage",
    "ET0_Temp", "Stress_Index",
    "Humidite_Amplitude", "Indice_Confort",
    "Mois_Plantation", "Annee_Plantation"
]

train.drop(columns=[c for c in drop_cols if c in train.columns], inplace=True)
test.drop(columns=[c for c in drop_cols if c in test.columns], inplace=True)

# -----------------------------
# 4️⃣ Colonnes numériques et catégorielles
# -----------------------------
num_features = [
    "Apport_N_(kg/ha)", "Apport_P_(kg/ha)", "Apport_K_(kg/ha)",
    "Age_plant", "Age_plant_sq", "Age_plant_cube", "Duree_Culture_Jours",
    "Total_NPK", "Temp_Amplitude", "Temp_Mean", "Temp_NPK",
    "N_P_ratio", "N_K_ratio", "P_K_ratio", "NPK_per_day", "Age_deviation"
]

if "Irrigation" in train.columns:
    num_features += ["Irrigation", "Irrigation_times_Age", "Irrigation_NPK"]

cat_cols = [
    "Campagne", "Region", "Code Ferme",
    "Libelle Site", "Libelle Variete", "Porte Greffe"
]
cat_cols = [c for c in cat_cols if c in train.columns]

train_cat = train.copy()
test_cat = test.copy()

for c in cat_cols:
    train_cat[c] = train_cat[c].astype(str).fillna("missing")
    test_cat[c] = test_cat[c].astype(str).fillna("missing")

# Fréquence des variétés
train_cat["Variete_freq"] = train_cat["Libelle Variete"].map(
    train_cat["Libelle Variete"].value_counts()
)
test_cat["Variete_freq"] = test_cat["Libelle Variete"].map(
    train_cat["Libelle Variete"].value_counts()
).fillna(1)

num_features.append("Variete_freq")

X_cb = train_cat[num_features + cat_cols]
X_test_cb = test_cat[num_features + cat_cols]

cat_features_idx = [X_cb.columns.get_loc(c) for c in cat_cols]

# -----------------------------
# 5️⃣ Log sur NPK
# -----------------------------
for col in ["Apport_N_(kg/ha)", "Apport_P_(kg/ha)", "Apport_K_(kg/ha)"]:
    X_cb[col] = np.log1p(X_cb[col])
    X_test_cb[col] = np.log1p(X_test_cb[col])

X_cb["NPK_Age"] = X_cb["Total_NPK"] * X_cb["Age_plant"]
X_test_cb["NPK_Age"] = X_test_cb["Total_NPK"] * X_test_cb["Age_plant"]

X_cb["Age_log"] = np.log1p(X_cb["Age_plant"])
X_test_cb["Age_log"] = np.log1p(X_test_cb["Age_plant"])

X_cb["NPK_per_month"] = X_cb["Total_NPK"] / (X_cb["Age_plant"] + 1)
X_test_cb["NPK_per_month"] = X_test_cb["Total_NPK"] / (X_test_cb["Age_plant"] + 1)

# -----------------------------
# 6️⃣ Cross-validation AMÉLIORÉ
# -----------------------------
rmse_scores = []
mse_scores = []
fold = 0

# Paramètres optimisés
params = {
  "iterations": 12000,
  "learning_rate": 0.006,
  "depth": 6, "l2_leaf_reg": 8,
  "bagging_temperature": 1.2,
  "random_strength": 2,
  "loss_function": "RMSE", "eval_metric": "RMSE",
  "od_type": "Iter",
  "od_wait": 700,
  "verbose": 300
}

# 5 folds au lieu de 3
gkf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_models = []

for tr_idx, val_idx in gkf.split(X_cb, y):
    model = CatBoostRegressor(**params, random_state=42+fold)
    model.fit(
        X_cb.iloc[tr_idx], y[tr_idx],
        cat_features=cat_features_idx,
        eval_set=(X_cb.iloc[val_idx], y[val_idx]),
        use_best_model=True,
        early_stopping_rounds=700
    )
    cv_models.append(model)

    val_pred = model.predict(X_cb.iloc[val_idx])
    rmse = root_mean_squared_error(y.iloc[val_idx], val_pred)
    mse = mean_squared_error(y.iloc[val_idx], val_pred)

    rmse_scores.append(rmse)
    mse_scores.append(mse)

    print(f"Fold {fold} | RMSE: {rmse:.4f} | MSE: {mse:.4f}")
    fold += 1

print(f"\n✅ RMSE CV moyen: {np.mean(rmse_scores):.4f}")
print(f"✅ MSE CV moyen: {np.mean(mse_scores):.4f}")

# -----------------------------
# 7️⃣ Prédiction avec clipping optimisé
# -----------------------------
final_pred = np.mean(
    [m.predict(X_test_cb) for m in cv_models],
    axis=0
)

# Clipping ajusté
final_pred = np.clip(
    final_pred,
    train[target].quantile(0.00),
    train[target].quantile(0.93)
)

submission = pd.DataFrame({
    "ID": test.index if "ID" not in test.columns else test["ID"],
    "Prediction_Rendement_T_ha": final_pred
})

submission.to_csv("submission_kaggle.csv", index=False)
print("✅ Submission créée")
```
